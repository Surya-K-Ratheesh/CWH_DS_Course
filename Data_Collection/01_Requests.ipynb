```python
import requests
```

```python
url = "https://books.toscrape.com/"
a = requests.get(url)
# a.text

with open (f"htmls/page1.html", "w") as f:
    f.write(a.text)
```

```python
res = a.status_code

if res == 200:
    print(f"{res} -> Sucess")
elif res == 404:
    print(f"{res} -> Not Found")
elif res == 403:
    print(f"{res} -> Forbidden")
elif res == 500:
    print(f"{res} -> Internal Server Error")
else:
    print("Unknown")
```

```python
#All Pages
for i in range(1, 51):
    a = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open(f"htmls/page{i}.html", "w", encoding = "utf-8") as f:
        f.write(a.text)
        print(f"Downloaded page {i} successfully")
```
