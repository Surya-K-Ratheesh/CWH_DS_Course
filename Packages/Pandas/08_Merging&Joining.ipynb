```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})
employees
```

```python
departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
departments
```

```python
pd.merge(employees, departments, on = "DeptID") #Inner Join
```

```python
pd.merge(employees, departments, on = "DeptID", how = "left") #Left Join -> Include all rows of employees
```

```python
pd.merge(employees, departments, on = "DeptID", how = "right") #Right Join -> Include all rows of departments
```

```python
pd.merge(employees, departments, on = "DeptID", how = "outer") #Outer Join -> Include all rows
```
