INHERITANCE

```python
class animal: #Superclass
    location = "Kannur"

    def __init__(self,name):
        self.name = name
    def speak(self):
        print("Generic Sound")

    def food(self):
        print("Pedigree")
```

```python
class Dog(animal): #inheritance
    def speak(self):
        print("Woof!")
```

```python
d = Dog("Bruno")
d.speak()
print(d.location)
```

```python
class Cat(animal):
    def speak(self):
        super().speak() #Using speak function of parent class
        print("Meow!")

    def food(self):
        super().food()
```

```python
c = Cat("Soman")
c.speak()
c.food()
print(c.location)
```

OPERATOR OVERLOADING

```python
class point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def sum(self, p):
        return point((self.x + p.x), (self.y + p.y))

    def print_point(self):
        return f"X is {self.x} & Y is {self.y}"

    def __add__(self, p):
        return point((self.x + p.x), (self.y + p.y))
```

```python
p1 = point(6, 2)
p2 = point(6, 3)

#p = p1.sum(p2)
p = p1 + p2 #__add__

p.print_point()
```

```python
class emp:
    company = "Dell"

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def print_details(self):
        print(f"Name: {self.name} & Age: {self.salary}")

    @staticmethod #No need of self - independent method
    def printtime():
        print("Time is now!")

    @classmethod
    def printClassDetails(cls):
        print(f"The company is {cls.company}")
```

```python
em = emp("SuryaK", 500000)
em.print_details()
em.printtime()
em.printClassDetails()
```
