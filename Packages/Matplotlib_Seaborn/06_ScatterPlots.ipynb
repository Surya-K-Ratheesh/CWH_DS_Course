```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 90, 55, 60, 25, 75, 85, 90]

colors = ['red' if score < 50 else 'green' for score in exam_scores]
sizes = [score * 2 for score in exam_scores]
```

```python
plt.scatter(study_hours, exam_scores, c = colors, s = sizes)
plt.grid()
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Marks Vs Study")
```

```python
plt.scatter(study_hours, exam_scores, c = exam_scores, cmap = 'viridis')
plt.colorbar(label = 'score')
plt.grid()
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Marks Vs Study")
```

```python
social_score = [12, 34, 67, 46, 76, 10, 76, 89, 77] #Colour dots can be used to represent something else too

plt.scatter(study_hours, exam_scores, c = social_score, cmap = 'viridis')
plt.colorbar(label = 'Social score')
plt.grid()
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Marks Vs Study")
```

```python
social_score = [12, 34, 67, 46, 76, 10, 76, 89, 77] #Colour dots can be used to represent something else too

plt.scatter(study_hours, exam_scores, c = social_score, cmap = 'viridis')
plt.colorbar(label = 'Social score')
plt.grid()
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Marks Vs Study")

for i in range(len(study_hours)):
    plt.annotate(f'S{i+1}', (study_hours[i] + 0.2, exam_scores[i])) #(f(what to write), (x,y))
```

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]
```

```python
plt.scatter(class_a_hours, class_a_scores, label = "Class A", color = 'blue')
plt.scatter(class_b_hours, class_b_scores, label = "Class B", color = 'red')

plt.title('Class A vs Class B')
plt.xlabel('Marks')
plt.ylabel('Hours of study')
plt.legend()
plt.grid()
plt.show()
```
