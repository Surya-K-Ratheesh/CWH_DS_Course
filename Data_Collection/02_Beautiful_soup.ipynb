```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
with open("htmls/page1.html") as f:
    content = f.read()
```

```python
soup = BeautifulSoup(content, "html.parser")
```

```python
#Inspected page1 and found name and cost are in h3
h3s = soup.find_all("h3")
h3s
```

```python
for h3 in h3s:
    print(h3.find("a")['title'])
```

```python
articles = soup.select("article.product_pod")
```

```python
items = []

for article in articles:
    title = article.find("h3").find("a")["title"]
    price = article.select_one("p.price_color").text.split("£")[1]
    rating_element = article.select_one("p.star-rating")
    rating = rating_element['class'][1]
    items.append([title, price, rating])

items
```

```python
df = pd.DataFrame(items, columns = ["Books", "Price", "Rating"])
df
```

```python
df.to_csv("data.csv", index = False)
```

```python
item = []

for i in range(1, 51):
    with open(f"htmls/page{i}.html", encoding = "utf-8") as f:
        contents = f.read()
        

    soups = BeautifulSoup(contents, "html.parser")
    
    articles1 = soup.select("article.product_pod")
    
    for article in articles1:
        title = article.find("h3").find("a")["title"]
        price = article.select_one("p.price_color").text.split("£")[1]
        
        rating_element = article.select_one("p.star-rating")
        rating = rating_element['class'][1]
        
        item.append([title, price, rating])

df1 = pd.DataFrame(item, columns = ["Books", "Price", "Rating"])

df1.to_csv("data_full.csv", index = False)
```
