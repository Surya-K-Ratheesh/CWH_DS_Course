```python
import numpy as np
```

```python
arr = np.array([23, 4, 3, 23])
arr.dtype
```

```python
arr.astype("float64")
```

```python
arr2 = np.array([23, 4, 3, 23], dtype = "float32")
arr2
```

```python
#Memory Usage
arr_int64 = np.array([1, 2, 3], dtype = "int64")
arr_int32 = np.array([1, 2, 3], dtype = np.int32)

print(arr_int64.nbytes) # 3 elements * 8 bytes
print(arr_int32.nbytes) # 3 elements * 4 bytes
```

```python
#String
arrs = np.array(['apple', 'banana', 'cherry'])
print(arrs.dtype)
```

```python
#Complex No.
arrc = np.array([1+2j, 3+4j, 5+6j])
print(arrc.dtype)
```

```python
arro = np.array([{'a': 1}, [1, 2, 3], 'hello'], dtype=object)
print(arro)
```
