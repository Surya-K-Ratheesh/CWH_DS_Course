```python
import numpy as np
import time
```

```python
l1 = [1, 2, 4]
l2 = [6, 7, 8]

list(zip(l1, l2))
```

```python
#Using Python Lists
size = 1_000_000

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x + y for x, y in zip(l1, l2)]
end = time.time()

print(f"Sum: {add[0:10]}")
print("Python list addition time: ",end - start)
```

```python
#Using NumPy Arrays
size = 1_000_000

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1 + l2
end = time.time()

print(f"Sum: {add[0:10]}")
print("NumPy Array addition time: ",end - start)
```

```python
n1 = np.array(list(range(10, 101)))
n2 = np.array(list(range(10, 101)))

start = time.time()
prod = n1 * n2
end = time.time()

print(f"Time taken: {end - start}")
```

```python
l1 = [x for x in range(1, 101)]
l2 = [x for x in range(1, 101)]

start = time.time()
prod = [x * y for x, y in zip(l1,l2)]
end = time.time()

print(f"Time taken: {end - start}")
```
