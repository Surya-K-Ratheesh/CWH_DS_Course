Fancy Indexing

```python
arr1 = np.array([1, 54, 23, 53, 2, 3, 34, 5, 6])
arr1[[1, 4, 6]]
```

```python
print(arr1[[1, 4, 6]])
```

Masking

```python
mask = arr1 > 25
print(arr1[mask])
```
