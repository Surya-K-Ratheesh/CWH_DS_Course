```python
import matplotlib.pyplot as plt
```

```python
activities = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
time_spent = [3, 2, 2, 5]
colours = ['blue', 'red', 'yellow', 'pink']
```

```python
plt.figure(figsize = (6, 6))
plt.pie(time_spent, labels = activities, colors = colours, startangle = 90, autopct = '%1.1f%%')
plt.axis() #Correct Aspect Ratio
```

STACKPLOTS

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]
 
labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['blue', 'red', 'yellow', 'pink']

plt.figure(figsize = (6, 6))
plt.stackplot(days, studying, playing, watching_tv, sleeping, labels = labels, colors = colours, alpha = 0.4)
plt.legend(loc = 'best')
plt.title("Time Distribution")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
```
